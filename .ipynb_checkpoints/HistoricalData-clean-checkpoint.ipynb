## 第五个版本  旬价的日期改成真实日期；删除错报的数据
把旬价的日期改成真实日期，如2月1日的旬价其实是1月下旬的价格，所以要把日期改成1月31日

```python
docAddress = r"D:\Data\2019价格监测系统\历史数据\旬度价格数据clean4补充空值小类.csv"
df = pd.read_csv(docAddress,encoding='UTF-8-SIG',engine='python')
df.head()
```

```python
#检查重复数据
sum(df.duplicated())
```

```python
df['taskdate'].value_counts()
```

删除20160211的26个数据、20180610的8个数据、20171230的4个数据，因为后面重新报了。删除前：1430983，删除后1430945

```python
#取反思路
df=df[~df['taskdate'].isin([20160211,20180610,20171230])]
```

```python
df.info()
```

```python
docAddress = r"D:\Data\2019价格监测系统\历史数据\clean5_把历史旬价的日期改成真实的日期-20191212.xlsx"
df_realdate = pd.read_excel(docAddress,sheet_name='realdate')
df_realdate.info()
```

```python
df_merge = pd.merge(df,df_realdate,how='inner',on='taskdate')
df_merge.info()
```

```python
df_merge.head()
```

删除taskdate列

```python
df_merge.drop(['记录数','taskdate'], axis=1,inplace=True) 
```

```python
df_merge.head()
```

把realdate改成taskdate

```python
df_merge.rename(columns={'realdate':'taskdate'},inplace=True) 
```

```python
df_merge.head()
```

```python
writer = r'D:\Data\2019价格监测系统\历史数据\旬度价格数据clean5改为真实日期-20191219.csv'
df_merge.to_csv(writer, index=False)
```

## 第六个版本根据v2列匹配现行三级分类名称
