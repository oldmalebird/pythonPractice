```python
#dataframe 二维表结构 ,Series (一行或者一列)
import pandas as pd
from pandas import Series,DataFrame
import numpy as np

# 创建Series对象并省略索引
'''
index 参数是可省略的，你可以选择不输入这个参数。
如果不带 index 参数，Pandas 会自动用默认 index 进行索引，类似数组，索引值是 [0, ..., len(data) - 1]
'''
sel =  Series([1,2,3,4])
print(sel)

# # 通常我们会自己创建索引
sel =  Series(data = [1,2,3,4], index = ['a','b','c','d'])

sel =  Series(data = [1,2,3,4], index = list('abcd'))
sel

# # 获取内容
print(sel.values)
# # 获取索引
print(sel.index)
# # 获取索引和值对
print(list(sel.iteritems()))
```

```python
# 将字典转换为Series
#json数据  

dict={"red":100,"black":400,"green":300,"pink":900}
se3=Series(dict)
print(se3)
print('--------------1-----------------')
# Series数据获取
sel =  Series(data = [1,2,3,4], index = list('abcd'))
result = list('abcd')
print(result)
print(sel)
print('--------------2-----------------')
# Series对象同时支持位置和标签两种方式获取数据
print('索引下标',sel['c'])
print('位置下标',sel[2])
print('--------------3-----------------')
# # # 获取不连续的数据
print('索引下标',sel[['a','c']])
print('位置下标',sel[[1,3]])
print('----------------4---------------')
# 可以使用切片或取数据
print('位置切片',sel[1:3])# 左包含右不包含
print('索引切片',sel['b':'d'])# 左右都包含
print('--------------5-----------------')
# # 重新赋值索引的值
#在原有数据基础上修改
sel.index = list('dcba')
print(sel)
print('--------------6-----------------')
# ReIndex重新索引,会返回一个新的Series(调用reindex将会重新排序，缺失值则用NaN填补)
#查询一下规则
#返回新的数据,不在原有数据修改
print(sel.reindex(['b','a','c','d','e']))
print('-------------7----------------')
# # Drop丢弃指定轴上的项()
se1=pd.Series(range(10,15))
print(se1)
print(se1.drop([2,3]))
```

```python
series1 = pd.Series([1,2,3,4],['London','HongKong','Humbai','lagos'])
series2 = pd.Series([1,3,6,4],['London','Accra','lagos','Delhi'])
print(series1)
print(series2)
print('--------------------------')
print(series1-series2)
print(series1+series2)
print(series1*series2)

# # 同样也支持numpy的数组运算
sel =  Series(data = [1,6,3,5], index = list('abcd'))
sel
print(sel[sel>3]) # 布尔数组过滤
print(sel*2) # 标量乘法
print(np.square(sel)) 
```

```python
# 1. 创建DataFrame
# 使用二维数组
df1 = DataFrame(np.random.randint(0,10,(4,4)),index=[1,2,3,4],columns=['a','b','c','d'])
print(df1)

# 使用字典创建(行索引由index决定，列索引由字典的键决定)
dict={
    'Province': ['Guangdong', 'Beijing', 'Qinghai', 'Fujian'],
    'pop': [1.3, 2.5, 1.1, 0.7],
    'year': [2018, 2018, 2018, 2018]}

df2=pd.DataFrame(dict)
print(df2)

# 使用from_dict
dict2={"a":[1,2,3],"b":[4,5,6]}
df6=pd.DataFrame.from_dict(dict2)
print(df6)

#索引相同的情况下，相同索引的值会相对应，缺少的值会添加NaN
data = {
    'Name':pd.Series(['zs','ls','we'],index=['a','b','c']),
    'Age':pd.Series(['10','20','30','40'],index=['a','b','c','d']),
    'country':pd.Series(['中国','日本','韩国'],index=['a','c','b'])
}

df = pd.DataFrame(data)
print(df)
# to_dict()方法将DataFrame对象转换为字典
dict = df.to_dict()
print(dict)
```

```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np

# dataframe常用属性
df_dict = {
    'name':['James','Curry','Iversion'],
    'age':['18','20','19'],
    'national':['us','China','us']
}
df = pd.DataFrame(data=df_dict,index=['0','1','2'])
print(df)
print('--------------------')
# 获取行数和列数
print(df.shape)

# # # 获取行索引
print(df.index.tolist())

# # # 获取列索引
print(df.columns.tolist())

# # 获取数据的类型
print(df.dtypes)

# 获取数据的维度
print(df.ndim)

# # values属性也会以二维ndarray的形式返回DataFrame的数据
print(df.values)

# # 展示df的概览
print(df.info())

# # 显示头几行,默认显示5行
print(df.head(2))

# # 显示后几行
print(df.tail(1))

# 获取DataFrame的列
print(df['name'])
# #因为我们只获取一列，所以返回的就是一个 Series
print(type(df['name']))

# # 如果获取多个列，那返回的就是一个 DataFrame 类型：
print(df[['name','age']])

print(type(df[['name','age']]))#
```

```python
#列表生成式:列表推导式:通俗理解就是使用for循环快速创建一个列表
my_list = []
for i in range(1,6):
    my_list.append(i)
print(my_list)
# 列表生成式的方式创建
my_list = [value*2 for value in range(1,6)]
print(my_list)

#统计每个元素的个数
my_list = [len(value) for value in ['abc',"ab"]]
print(my_list)
my_list = [value+'haha' for value in ['abc',"ab"]]
print(my_list)

#双层for循环 
my_list = [[x,y]for x in range(1,3) for y in range(1,3)]
print(my_list)
for x in range(1,3):
    print(x)
    for y in range(1,3):
        print(y)

# #结合 if 语句 
my_list = [value for value in range(1,11)if value % 2 == 0]
print(my_list)
```

```python
#获取最高温度
#loc自定索引切片方式
data.loc[:,'最高温度']

#改变数据的行索引返回一个新的数据
print(data.set_index('日期'))
#inplace = True在原有数据基础上修改
data.set_index('日期',inplace=True)
data
data.index.name = None
print(data)
data.head()

#将最高温度改变成数字
data.loc[:,'最高温度']=data['最高温度'].str.replace('℃','').astype('int32')
data
# #将最低温度修改成数字
data.loc[:,'最低温度']=data['最低温度'].str.replace('℃','').astype('int32')
data
# #全年最高温度那一天的信息,ascending=True 升序
result = data.sort_values(by='最高温度',ascending=True)
#进行数据预处理 不满足分析挖掘的条件
result

#获取一行 
#iloc 位置索引
result = result.iloc[0,:]
print(result)

#获取某几行数据某几列
#[2019/1/3 1/4/ ]
#自定义索引
# data.head()
data.loc[['2019/1/3','2019/1/4','2019/1/5'],['最高温度','最低温度']]

# data.head()
#最高温度大于30度


result = data.loc[data['最高温度']>32]
result = data.loc[(data['最高温度']<30)&(data['最低温度']>20)]
result

#查询9月的所有数据
# 所有数据取出来 2019/09 startswith
data.head()
index_list = data.index.tolist()
# #i.startswith('2019/09' -> TRUE fasle
result = [i.startswith('2019/9') for i in index_list]

result = data.loc[result,:]
result
#[if i == 2019/9 ->ELSE False for i in index_list]
#列表生成式[for ]for后面只能写if,不能写else

#温差一列 
data.loc[:,'温差']=data['最高温度']-data['最低温度']
# data

#温差>=10 大温差 其他情况小温差 
#增加新的列温差类型
data.loc[:,"温差类型"]=['大温差' if value>=10 else '正常' for value in data['温差']]
data
# #全年里面有多少天是正常温差
data = data.loc[:,'温差类型'].value_counts()

data = data.loc[:,'风向'].value_counts()
data
# #默认按照第一个参数排序,再去按照温差排序
data.sort_values(by=['最高温度','温差'],ascending=False)
data.sort_values(by=['最高温度','温差'],ascending=[False,True])
```

```python
import pandas as pd 
import numpy as np 
#空值处理
#删除缺失值 dropna 
# filna 填充 
#isnull 
#headers = 2 将第二行作为列索引
#headers = None 避免读取的时候将第一行作为索引
#行索引
#跳跃到索引为2的行上
df = pd.read_excel('./student_excel.xlsx',header=2)
# df 
#如果你想要截取excle的一段范围skiprows
df = pd.read_excel('./student_excel.xlsx',skiprows=2,usecols="B:E")
df 
#删除全是空值的列
#how,= any只要有空值就删除 all 全是空值删除
df.dropna(axis=0,how='all',inplace=True)
print(df)

# #
df.dropna(axis=0,how='any',inplace=True)
df
#填充的方式
#分数参数是列索引
df.fillna({'姓名':"无名氏"},inplace=True)
df
```

```python
# 使用join合并，着重关注的是行的合并
import pandas as pd
df3=pd.DataFrame({'Red':[1,3,5],'Green':[5,0,3]},index=list('abc'))
df4=pd.DataFrame({'Blue':[1,9,8],'Yellow':[6,6,7]},index=list('cde'))
print(df3)
print(df4)

# 简单合并（默认是left左连接,以左侧df3为基础）
# df3.join(df4,how='left')

# # 右链接
# df3.join(df4,how='right')

# # 外链接
# df3.join(df4,how='outer')

# # 合并多个DataFrame对象
df5=pd.DataFrame({'Brown':[3,4,5],'White':[1,1,2]},index=list('aed'))
print(df5)
# #课下练习尝试多个表 使用右链接的方式 
# #为什么告知我
df3.join([df4,df5],how='right')
```
