```python
import pandas as pd
data = pd.read_csv('./BJ_tianqi.csv')
```

```python
# 改变数据的行索引。将行索引设置为日期
new_data = data.set_index('日期')
new_data

new_data.index.name = None
```

```python
new_data
```

```python
new_data.loc[:,'最高温度'] = new_data['最高温度'].replace('℃', '').astype('int64')
new_data.loc[:,'最低温度'] = new_data['最低温度'].replace('℃', '').astype('int64')
```

```python
new_data.loc[:,'最高温度'] = new_data['最高温度'].str.replace('℃', '').astype('int64')
new_data.loc[:,'最低温度'] = new_data['最低温度'].str.replace('℃', '').astype('int64')
```

```python
new_data
```

```python
reslut = new_data.sort_values(by='最高温度', ascending=False)
```
