## 批量读取相同内容

```python
root_path = r"D:\huodian\huodianDataMerge\9月份\#5"
files = os.listdir(root_path) 
n = len(files)
print(files)
print(n)
```

```python
df_full =  pd.read_excel(root_path + "\\"+ files[0],names=['颗粒物','SO2','NOX'],usecols="F,I,L",skiprows=6,nrows=24)
print(df_full)
```

```python
for i in range(1,n):
    fullFilePath = root_path + "\\"+ files[i]
    df = pd.read_excel(fullFilePath,names=['颗粒物','SO2','NOX'],usecols="F,I,L",skiprows=6,nrows=24)
    print(df)
    df_full = pd.concat([df_full,df],sort=False)
    print(df_full)
writer = pd.ExcelWriter(r"D:\huodian\huodianDataMerge\9月份#5.xlsx")
df_full.to_excel(writer,'sheet1',index=False)
writer.save()    
```
